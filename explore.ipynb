```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_excel('./src/Boston_Housing.xlsx')
```

# Beginning discover the dataset

```python
df.info()
```

```python
# show the first 5 rows
df.head()
```

```python
# show the last 5 rows
df.tail()
```

```python
# Put columns names in lowercase
df.columns = df.columns.str.lower()
df
```

```python
# Rename columns
df.rename(columns={'crim': 'crime', 'zn': 'zone', 'chas': 'river', 'rm': 'room', 'dis': 'district', 'b': 'black_population', 'medv': 'price'}, inplace=True)
```

```python
# Let's count the null values
df.isnull().sum()
```

# Unique values

```python
# Let's count the unique values in each column
df.nunique()
```

# First analysis

```python
df.describe()
```

```python
# let's saw the correlation between the variables
df.corr()["price"].sort_values(ascending=False)
```

```python
# Regplots with price and room
sns.regplot(x="room",y="price", data=df, fit_reg=True)
plt.show()
```

```python
sns.regplot(x="lstat",y="price", data=df, fit_reg=True)
plt.show()
```

```python
sns.scatterplot(x='ptratio', y='price', data=df)
plt.show()
```
